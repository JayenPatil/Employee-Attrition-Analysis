# **FEATURE ENGINEERING**

To simplify analysis and improve data interpretation, new features were created from existing variables.

- **Age_Group** categorizes employees into age ranges.
- **Income_Group** groups employees based on their monthly income.
- **Experience_Group** classifies employees by years spent at the company.
- **Overtime_Status** provides more descriptive labels for overtime information.

These engineered features can support future analyses and help segment employees into meaningful groups for business decision-making.

```python
#Feature Engineering
#Age Group
df['Age_Group'] = pd.cut(df['Age'],bins=[18,30,40,50,60],labels=['18-30','31-40','41-50','51-60'])
```

```python
#Income Group
df['Income_Group'] = pd.cut(df['MonthlyIncome'],bins=[0,3000,6000,10000,20000],labels=['Low','Medium','High','Very High'])
```

```python
#Year at Company Group
df['Experience_Group'] = pd.cut(df['YearsAtCompany'],bins=[0,2,5,10,40],labels=['0-2 Years','3-5 Years','6-10 Years','10+ Years'])
```

```python
# Overtime Category
df['Overtime_Status'] = df['OverTime'].map({'Yes':'Work Overtime','No': 'No Overtime'})
```

```python
df[['Age_Group','Income_Group','Experience_Group','Overtime_Status']].head()
```

# **KPI SECTION**

Key Performance Indicators (KPIs) provide a quick overview of the workforce before detailed analysis.

The following KPIs summarize the total number of employees, active employees, employees who left the organization, attrition rate, average monthly income, average employee age, and average years at the company.

## INSIGHT

The organization has a workforce of **1,470 employees**, of which **1,233 are active** and **237 have left**, resulting in an **attrition rate of 16.12%**.

On average, employees are **36.9 years old**, earn approximately **6,503** per month, and have **7 years** of tenure with the organization. These KPIs provide an overall snapshot of the workforce before the detailed exploratory data analysis.

## **DEPARTMENT-WISE EMPLOYEE ATTRITION ANALYSIS**

## **CORRELATION HEATMAP**

This analysis examines the correlation between important numerical features in the dataset. Correlation analysis helps identify whether two variables move together positively, negatively, or have little to no relationship.

The correlation coefficient ranges from -1 to +1:

- +1 → Strong Positive Correlation
- 0 → No Correlation
- -1 → Strong Negative Correlation


---


**INSIGHT**

The heatmap shows several meaningful relationships among employee characteristics:

- Monthly Income and Total Working Years have a strong positive correlation (0.77), indicating that employees with more work experience generally earn higher salaries.
- Age and Total Working Years also have a strong positive correlation (0.68), suggesting that older employees tend to have more overall work experience.
- Years at Company has a moderate positive correlation with Total Working Years (0.63), meaning employees with more total experience often spend more years with the organization.
- Monthly Income and Years at Company show a moderate positive correlation (0.51), indicating that employees with longer tenure generally receive higher salaries.
- Job Satisfaction and Work-Life Balance have very weak correlations with the selected numerical variables, suggesting that these factors are influenced by aspects other than age, salary, or experience.

```python
# Correlation Heatmap
plt.figure(figsize=(10,6))
corr_cols = ['Age','MonthlyIncome','YearsAtCompany','TotalWorkingYears','JobSatisfaction','WorkLifeBalance']
sns.heatmap(df[corr_cols].corr(),annot=True,fmt='.2f',cmap='coolwarm',linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=15,fontweight='bold')
plt.tight_layout()
plt.show()
```
